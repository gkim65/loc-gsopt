```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import numpy as np

def plot_south_pole(coords):
    plt.figure(figsize=(8, 8))
    
    # Create a Basemap centered on the South Pole
    m = Basemap(projection='spstere', boundinglat=-80, lon_0=180, resolution='l')
    
    # Draw map details
    m.drawcoastlines()
    m.drawparallels(np.arange(-90., -60., 5.), labels=[False, False, False, False])
    m.drawmeridians(np.arange(-180., 180., 30.), labels=[True, True, True, True])
    
    # Convert coordinates to the map projection
    lons, lats = zip(*coords)
    x, y = m(lons, lats)
    
    # Plot the points
    m.scatter(x, y, marker='o', color='red', edgecolors='black', s=50, label="Coordinates")
    
    plt.legend()
    plt.title("South Pole View of Given Coordinates")
    plt.show()

# Example usage
coords = [
    [92.44410570135645, -90.0], [-73.01224079830891, -89.95426045890125],
    [-4.078723407808447, -89.95514937821778], [-20.501266626711264, -89.91355093899406],
    [-159.10268897441438, -89.95438003369789], [-115.71665466969782, -89.93402443627843],
    [-48.87503465712989, -89.90572106322753], [-180.0, -89.91518762080831],
    [-149.7087766419795, -89.91006098768894], [-77.3218705116782, -89.90973730577142]
]
coords2 = [[92.44410570135645, -90.0], [80.29121758653447, -85.52335783477325], [20.23246326417585, -85.52333185018881]]

plot_south_pole(coords2)
```
