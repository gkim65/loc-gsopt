```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np

# Average runtimes in seconds for each case
runtimes_seconds = [
    842,      # 1 ground station
    6811,     # 2 ground stations
    22091,    # 3 ground stations
    121390    # 4 ground stations (computed above)
]
labels = ['1 Ground Station', '2 Ground Stations', '3 Ground Stations', '4 Ground Stations']
runtimes_hours = [t / 3600 for t in runtimes_seconds]

x = np.arange(len(labels))
width = 0.6

fig, ax = plt.subplots()
bars = ax.bar(x, runtimes_hours, width, color=['blue', 'green', 'orange', 'maroon'])

ax.set_ylabel('Average Runtime (hours)')
ax.set_title('Average Runtime by Number of Ground Stations')
ax.set_xticks(x)
ax.set_xticklabels(labels)

for bar in bars:
    height = bar.get_height()
    ax.annotate(f'{height:.2f}',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3),
                textcoords='offset points',
                ha='center', va='bottom')

plt.show()

```
